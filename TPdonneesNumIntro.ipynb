Pour chaque valeure entière et impaire de x, tracez un petit segment qui passe par ce point (xi,yi) et de pente exp(xi). A quoi cela correspond?

```python
x = np.linspace(0, 10, 100)
y = np.exp(x)

plt.plot(x, y, label='f(x) = e^x')

# Ajouter des segments pour chaque valeur entière et impaire de x entre 0 et 10
for xi in range(1, 10, 2):
    yi = np.exp(xi)
    pente = np.exp(xi)
    # Calculer les points du segment
    x_segment = np.array([xi-0.5, xi+0.5])
    y_segment = yi + pente * (x_segment - xi)
    plt.plot(x_segment, y_segment, 'r--')

plt.xlabel('x')
plt.ylabel('exp(x)')
plt.title('Tracé de la fonction exp(x) avec segments de pente exp(xi)')
plt.legend()
plt.show()
```

#### logarithme

Explorez de la même façon la fonction logarithme. Jouez avec les valeurs de x. Est-ce un logarithme népérien ou en base 10?

```python
# dessin 1
x = np.linspace(0.1,10,100)
y = np.log(x)
xi = np.exp(1)
yi = np.log(xi)
plt.plot(xi, yi, 'ro')  
plt.text(xi, yi, f'({xi}, {yi:.2f})', fontsize=12, verticalalignment='bottom')

plt.xlabel('x')
plt.ylabel('log(x)')
plt.title('Tracé de la fonction log(x)')
plt.plot(x,y)
```

Ajoutez les segments de pente 1/xi:

```python
x = np.linspace(0.1,10,100)
y = np.log(x)
plt.plot(x, y, label='f(x) = log(x)')


for xi in range(1, 2, 10):
    yi = np.log(xi)
    pente = np.log(xi)
    # Calculer les points du segment
    x_segment = np.array([1/(xi-0.5),1/(xi+0.5)])
    y_segment = yi + pente * (x_segment - xi)
    plt.plot(x_segment, y_segment, 'r--')

plt.xlabel('x')
plt.ylabel('log(x)')
plt.title('Tracé de la fonction log(x)')
plt.plot(x,y)
```
